```python
To=ToQ(P) # torque in tha system is cero
```

```python
To.set(alpha1=alpha,alpha2=alpha+2*pi/2,alpha3=alpha+4*pi/3,m1=m,m2=3*m,m3=2*m)
```

```python
To.texpand()  # expand To , of course we get solve in one steep but is more lindo asi 
```

```python
To.reduFac() 
```

```python
a=To.solve(alpha,'a') 
```
