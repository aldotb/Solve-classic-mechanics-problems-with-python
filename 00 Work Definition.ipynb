A block of mass $m$ is initially at rest:

$$
v_1 = 0

at position

x_1 = 0

###  Create P whit initial condition x=0,v1=0,y1=0,y2=0

```python
P=mparticle(x1=0,v1=0,y1=0,y2=0)
```

###  Add forces acting on P

Component of gravity along the plane:

- Weight   : $mg$
- F vector : $F$ 
- Friction : $fr$
- Normal   : $N1$ 

```python
# P.add_forza(value,angle,posx,posy), posx,posy in some cases
P.add_forza(m*g,-pi/2)
P.add_forza(F,alpha)
P.add_forza(fr,pi)
P.add_forza(N1,pi/2)
```

### How work F in the problem

Horizontal component:

$$
F_x = F\cos\alpha
$$

Vertical component:

$$
F_y = F\sin\alpha
$$


### Vertical Dynamics (No Vertical Motion)

Since the block does not move vertically:

$$
\sum F_y = 0
$$

$$
N + F\sin\alpha - mg = 0
$$

Normal force:

$$
\boxed{
N = mg - F\sin\alpha
}
$$

Condition to maintain contact:

$$
mg > F\sin\alpha
$$


###  Resultant Forces Along y axes = cero

```python
Fy=P.Fy('Fy')
```

```python
# solve N1 and N1 is MyEq object :N1=Fy.solve('N1'), include 'N'
# solve N1 and get only the value  :N1=Fy.solve(N1), not include '' 
N1=Fy.solve('N1')  
```

### Horizontal Dynamics

Newton’s Second Law:

$$
\sum F_x = ma
$$

$$
F\cos\alpha = ma
$$

Acceleration:

$$
\boxed{
a = \frac{F\cos\alpha}{m}
}
$$


```python
Fx=P.Fx('Fx')
```

```python
Q=MQ(Fx,m*a)
```

```python
Q.set(fr=mu*N1())
```

```python
Q.alone(a)
```

### Work–Energy Verification

Work done by the applied force:

$$
W = F_x x_2
$$

$$
W = F\cos\alpha \, x_2
$$

Work–Energy Theorem:

$$
W = \Delta K
$$

$$
F\cos\alpha \, x_2 =
\frac{1}{2} m v_2^2
$$

Solving:

$$
v_2 =
\sqrt{
\frac{2F\cos\alpha}{m} \, x_2
}
$$

Consistent with Newton’s law ✔
