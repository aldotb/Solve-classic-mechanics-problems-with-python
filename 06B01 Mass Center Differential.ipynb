## 1️⃣ Sistema de coordenadas

Ponemos el origen en el **centro del círculo**, con:

- eje $x$ horizontal
- eje $y$ vertical
- la semicircunferencia en $y \ge 0$

Usamos coordenadas polares:

$$
x = r\cos\theta, \qquad y = r\sin\theta
$$

con:

$$
0 \le r \le R, \qquad 0 \le \theta \le \pi
$$

El elemento de área es:

$$
dA = r\,dr\,d\theta
$$

---

## 2️⃣ Masa total

$$
M = \iint \sigma(r)\, dA
= \int_0^\pi \int_0^R c r^2 \cdot r \, dr \, d\theta
$$

$$
M = c \int_0^\pi d\theta \int_0^R r^3 dr
$$

$$
M = c\pi \left(\frac{R^4}{4}\right)
= \frac{c\pi R^4}{4}
$$

```python
M=My2Intg(c*r*r*r,'M',(r,0,R),(alpha,0,pi))
```

```python
M.doit()
```

## 3️⃣ Momento respecto al eje $x$

$$
\bar y = \frac{1}{M} \iint y\, \sigma(r)\, dA
$$

Como $y = r\sin\theta$:

$$
\iint y\,\sigma\,dA
= \int_0^\pi \int_0^R (r\sin\theta)(c r^2)(r\,dr\,d\theta)
$$

$$
= c \int_0^\pi \sin\theta\, d\theta \int_0^R r^4 dr
$$

$$
= c(2)\left(\frac{R^5}{5}\right)
= \frac{2cR^5}{5}
$$

```python
y=My2Intg(r*sin(alpha)*c*r*r*r,'y',(r,0,R),(alpha,0,pi))
```

```python
y.doit()
```

## 4️⃣ Centro de masas

$$
\bar y
= \frac{\frac{2cR^5}{5}}{\frac{c\pi R^4}{4}}
= \frac{8R}{5\pi}
$$
