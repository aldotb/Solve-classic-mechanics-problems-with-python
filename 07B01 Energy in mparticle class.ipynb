### Example
- An object slides down an inclined plane with angle alpha and H height
- When it reaches the ground its speed is V and due to friction the object slows down until it stops.
- Find L

```python
from lib_Physics import *
```

```python
# create Physiscs mparticle object A to B
P=mparticle(y1=H,y2=0,v1=0,v2=V)
```

```python
# Energy A to B equal cero
Q1=MyEqEq(P.energy('KP'),0) # MyEqEq = MQ(a,b) : return a = b equation
```

```python
Q1.alone(V*V)
```

```python
Q1.Rpow() # Rpow() : Sqrt, Rpow(n) = expr ** 1/n , Rpow(a,b) = expr ** a/b 
```

```python
Q1.rsimplify() # rsimplify kill sqrt(pow)
```
