```python
from Libaldomath import *
```

```python
A=mparticle(x1=0,v1=40,a=pi/2,y1=0,g=0)
```

```python
B=mparticle(x1=20,y1=0,v1=30,a=pi/2+pi/3,g=0)
```

```python
Pa=Point2D(A.xpos(),A.ypos())
```

```python
Pb=Point2D(B.xpos(),B.ypos())
```

```python
Pa.distance(Pb)
```

```python
Vr=MyEq(Pa.distance(Pb),'Vr',var=t)
```

```python
Vr.expand()
```

```python
dV=Vr.diff('dV')
```

```python
dV.reducecero()
```

```python
t1=dV.solve('t')
```

```python
Vr.set(t1)
```
