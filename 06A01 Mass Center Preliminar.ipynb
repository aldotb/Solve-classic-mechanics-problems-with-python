# Center of Mass of a Horizontal Rod

---

## 1. Definition of Center of Mass

The **center of mass** of a one-dimensional object along the x-axis is defined as:

$$
x_\mathrm{cm} = \frac{\text{total moment}}{\text{total mass}} = \frac{\int_0^L x \, dm}{\int_0^L dm}
$$

where \(dm\) is an infinitesimal element of mass.

---

```python
Xm = MyIntg(x,'Xm',var=m,x1=0,x2=L)
M = MyIntg(1,'M',var=m,x1=0,x2=L)
```

## 2. Homogeneous Rod

If the rod has **length \(L\)** and **mass \(M\)** and is **homogeneous**, its linear density is:

$$
\lambda = \frac{M}{L} \quad \Rightarrow \quad dm = \lambda \, dx
$$

Substituting into the center of mass formula:

$$
x_\mathrm{cm} = \frac{\int_0^L x \lambda \, dx}{\int_0^L \lambda \, dx}
$$

Factoring out \(\lambda\) from the integrals:

$$
x_\mathrm{cm} = \frac{\lambda \int_0^L x \, dx}{\lambda \int_0^L dx} = \frac{\int_0^L x \, dx}{\int_0^L dx}
$$

---

```python
Xm.changediff(x,lamda*dx)
M.changediff(x,lamda*dx)
```

## 3. Calculating the Integrals

$$
\int_0^L x \, dx = \frac{L^2}{2}, \quad \int_0^L dx = L
$$

Therefore:

$$
x_\mathrm{cm} = \frac{L^2/2}{L} = \frac{L}{2}
$$

---

```python
Xm.doit()
```

```python
M.doit()
```

## 4. Conclusion

The **center of mass** of a homogeneous rod is at **the midpoint** of the rod:

$$
\boxed{x_\mathrm{cm} = \frac{L}{2}}
$$

---

## 5. Note on Variable Density

If the rod is not homogeneous and has a variable density \(\lambda(x)\), the general formula is:

$$
x_\mathrm{cm} = \frac{\int_0^L x \, \lambda(x) \, dx}{\int_0^L \lambda(x) \, dx}
$$
