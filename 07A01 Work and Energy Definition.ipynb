## 2. Definition of Work

The differential work done by a force is defined as:

$$
dW = \vec{F} \cdot d\vec{r}
$$

For one-dimensional motion:

$$
dW = F(x)\, dx
$$

Total work between $x_1$ and $x_2$:

$$
W = \int_{x_1}^{x_2} F(x)\, dx
$$

## 3. Work–Energy Theorem

Newton’s Second Law:

$$
F = ma
$$

Using:

$$
a = \frac{dv}{dt}
\quad \text{and} \quad
v = \frac{dx}{dt}
$$

We write:

$$
F = m \frac{dv}{dt}
$$

Multiply both sides by $v$:

$$
F v = m v \frac{dv}{dt}
$$

But:

$$
v dt = dx
$$

Therefore:

$$
F dx = m v dv
$$

Integrating:

$$
\int_{x_1}^{x_2} F(x)\, dx
=
\int_{v_1}^{v_2} m v \, dv
$$

Right-hand side:

$$
\int m v\, dv
=
\frac{1}{2} m v^2
$$

Thus:

$$
W =
\frac{1}{2} m v_2^2
-
\frac{1}{2} m v_1^2
$$

## 4. Kinetic Energy

We define kinetic energy:

$$
K = \frac{1}{2} m v^2
$$

Therefore:

$$
W = \Delta K
$$

This is the Work–Energy Theorem.

---

## 5. Conservative Forces and Potential Energy

If the force depends only on position and satisfies:

$$
W = -\Delta U
$$

We define potential energy:

$$
U(x) = - \int F(x)\, dx
$$

Then total mechanical energy is:

$$
E = K + U
$$

---

## 6. Conservation of Mechanical Energy

If only conservative forces act:

$$
W_{\text{non-conservative}} = 0
$$

Then:

$$
K_1 + U_1 = K_2 + U_2
$$

Or equivalently:

$$
E = \text{constant}
$$

---

## 7. Special Case: Constant Force

If $F$ is constant:

$$
W = F (x_2 - x_1)
$$

Then:

$$
F (x_2 - x_1)
=
\frac{1}{2} m v_2^2
-
\frac{1}{2} m v_1^2
$$

If $v_1 = 0$ and $x_1 = 0$:

$$
v_2 =
\sqrt{
\frac{2F x_2}{m}
}
$$

---

## 8. Interpretation

- Work transfers energy.
- Conservative forces store energy as potential energy.
- The Work–Energy Theorem connects Newton’s law with energy methods.

$$

$$

$$

Normal force:

$$
\boxed{
N = mg - F\sin\alpha
}
$$

Condition to maintain contact:

$$
mg > F\sin\alpha
$$
