This section defines the main variables used in motion simulations and physics problems.

* declare Physic Object:

````python
P=mparticle()
````

## 🔹 Basic Physical Quantities

| Variable | Meaning              |
| -------- | -------------------- |
| `m`      | Mass                 |
| `g`      | Gravity acceleration |
| `v`      | Velocity             |
| `v1`     | Initial velocity     |
| `v2`     | Final velocity       |

---

## 🔹 Position Variables

| Variable | Meaning                             |
| -------- | ----------------------------------- |
| `x1`     | Initial position (x)                |
| `y1`     | Initial height / initial y position |
| `x2`     | Final position (x)                  |
| `y2`     | Final position (y)                  |

---

## 🔹 Linear Acceleration

| Variable   | Meaning                     |
| ---------- | --------------------------- |
| `a` / `ac` | Acceleration                |
| `ax`       | Acceleration in x direction |
| `ay`       | Acceleration in y direction |
| `at`       | Tangential acceleration     |

Note:

In **projectile motion**:

$$
a_x = 0
$$

$$
a_y = -g
$$

---

## 🔹 Circular Motion Variables

| Variable | Meaning                   |
| -------- | ------------------------- |
| `r`      | Radius of circular motion |
| `r1`     | Initial radius            |
| `r2`     | Final radius              |
| `w1`     | Initial angular velocity  |
| `w2`     | Final angular velocity    |
| `aw`     | Angular acceleration      |

---

## 🔹 Angle

| Variable | Meaning                                  |
| -------- | ---------------------------------------- |
| `ang`    | Launch angle (used in projectile motion) |

---

## 🔹 Example Initialization in Python

````python
P=mparticle()
P.m = m      # mass
P.g = g      # gravity

# Initial position
P.x1 = x1
P.y1 = y1

# Final position
P.x2 = x2
P.y2 = y2

# Velocities
P.v = v
P.v1 = v1
P.v2 = v2

# Accelerations
P.ac = ac
P.ax = ax
P.ay = ay
P.at = at

# Circular motion
P.r = r
P.r1 = r1
P.r2 = r2

# Angular motion
P.aw = aw
P.w1 = w1
P.w2 = w2

# Angle
P.ang = 0
````

 

# 📌 Particle Position Functions

In **`lib_Physics`**, a particle object can be created using:

```python
P = mparticle()
```

This object represents a **moving particle** with properties like position, velocity, and acceleration.

---

# 📌 `P.xpos(t)`

Returns the **x-coordinate position** of the particle as a function of time.

### Syntax

```python
P.xpos(t)
```

### Parameters

| Parameter | Meaning       |
| --------- | ------------- |
| `t`       | time variable |

### Returns

The position in the **x direction** at time `t`.

### General kinematic equation

If acceleration is constant:

$$
x(t) = x_0 + v_{x0} t + \frac{1}{2} a_x t^2
$$

Where:

| Symbol   | Meaning               |
| -------- | --------------------- |
| $x_0$    | initial x position    |
| $v_{x0}$ | initial velocity in x |
| $a_x$    | acceleration in x     |
| $t$      | time                  |

### Example

```python
P = mparticle()

P.x1 = 0      # initial position
P.vx = 10     # initial velocity
P.ax = 2      # acceleration

P.xpos(t)
```

---

# 📌 `P.ypos(t)`

Returns the **y-coordinate position** of the particle as a function of time.

### Syntax

```python
P.ypos(t)
```

### Parameters

| Parameter | Meaning       |
| --------- | ------------- |
| `t`       | time variable |

### Returns

The position in the **y direction** at time `t`.

### General kinematic equation

$$
y(t) = y_0 + v_{y0} t + \frac{1}{2} a_y t^2
$$

Where:

| Symbol   | Meaning               |
| -------- | --------------------- |
| $y_0$    | initial y position    |
| $v_{y0}$ | initial velocity in y |
| $a_y$    | acceleration in y     |
| $t$      | time                  |

---

# 📌 Example: Projectile Motion

```python
P = mparticle()

P.x1 = 0
P.y1 = 0

P.vx = 10
P.vy = 15

P.ax = 0
P.ay = -9.81

P.xpos(t)
P.ypos(t)
```

Trajectory equations:

$$
x(t) = v_x t
$$

$$
y(t) = v_y t - \frac{1}{2}gt^2
$$
