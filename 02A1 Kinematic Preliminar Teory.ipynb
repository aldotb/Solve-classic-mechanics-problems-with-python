# 📌 2.1 What is Kinematics?

**Kinematics** is the branch of mechanics that studies **motion** without considering the forces that cause it.

It describes:

* **Position**
* **Velocity**
* **Acceleration**
* **Time**

Basic variables:

 Variables in mparticle() class        
 
          
 - $x1$    Initial position
 - $x2$    Final position
 - $v1$    Initial Velocity           
 - $v2$    Final velocity  
 - $ac$    Acceleration      
 - $t$     Time              
 - $w1$    Initial angular velocity
 - $w2$    Final angular velocity
---

# 📌 2.2 Displacement

**Displacement** is the change in position.

$$
\Delta x = x - x_0
$$

Where:

* $x1$ = final position
* $x2$ = initial position

Displacement is a **vector quantity**.

---

# 📌 2.3 Average Velocity

Average velocity is defined as:

$$
v_{avg} = \frac{\Delta x}{\Delta t}
$$

Where:

$$
\Delta t = t - t_0
$$

---

# 📌 2.4 Uniform Motion (MRU)

**Uniform Motion** means the velocity is **constant**.

This implies:

$$
a = 0
$$

The position as a function of time is:

$$
x(t) = x_0 + vt
$$

Where:

* $x_0$ = initial position
* $v$ = constant velocity
* $t$ = time

Properties:

* Velocity is constant
* Acceleration is zero
* Position changes linearly with time

---

# 📌 2.5 Uniformly Accelerated Motion (MRUV)

In this motion, the **acceleration is constant**.

### Velocity equation

$$
v = v_0 + at
$$

---

### Position equation

$$
x = x_0 + v_0 t + \frac{1}{2}at^2
$$

---

### Velocity–position relation

$$
v^2 = v_0^2 + 2a(x - x_0)
$$

These equations are called the **kinematic equations**.

---

# 📌 2.6 Acceleration

Acceleration is the rate of change of velocity.

$$
a = \frac{\Delta v}{\Delta t}
$$

Where:

$$
\Delta v = v - v_0
$$

Units:

$$
[m/s^2]
$$

---

# 📌 2.7 Free Fall

**Free fall** is motion under the influence of **gravity only**.

The acceleration is constant:

$$
g \approx 9.81 , m/s^2
$$

Directed **downward**.

---

### Velocity in free fall

$$
v = v_0 + gt
$$

---

### Position in free fall

$$
y = y_0 + v_0 t + \frac{1}{2}gt^2
$$

---

### Velocity–height relation

$$
v^2 = v_0^2 + 2g(y - y_0)
$$

---

# 📌 2.8 Special Case: Dropping an Object

If an object is **dropped from rest**:

$$
v_0 = 0
$$

Then:

Velocity:

$$
v = gt
$$

Position:

$$
y = y_0 + \frac{1}{2}gt^2
$$

---

# 📌 2.9 Graphs in Kinematics

### Position vs Time

* MRU → straight line
* MRUV → parabola

---

### Velocity vs Time

* MRU → constant line
* MRUV → linear function

---

### Acceleration vs Time

* MRU → zero
* MRUV → constant value
