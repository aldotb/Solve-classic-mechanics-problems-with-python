# Statics – Basic Theory

## 1. Definition of Statics

Statics is the branch of classical mechanics that studies **forces acting on bodies in equilibrium**, meaning that the body has **no acceleration**.

If the acceleration of a system is zero:

$$
\vec{F}_{net} = m\vec{a} = 0
$$

Therefore, the vector sum of all forces acting on the body must be zero.

---

## 2. Force

A **force** is a push or pull acting on a body.

It is a **vector quantity**, meaning it has:

- Magnitude
- Direction
- Point of application

---

## 3. First Condition of Equilibrium

The sum of all forces must be zero.

$$
\sum \vec{F} = 0
$$

In Cartesian coordinates:

$$
\sum F_x = 0
$$

$$
\sum F_y = 0
$$

$$
\sum F_z = 0
$$

---

## 4. Second Condition of Equilibrium

The sum of torques must be zero.

$$
\sum \tau = 0
$$

---

## 5. Torque

Torque measures the tendency of a force to rotate an object.

$$
\tau = rF\sin\theta
$$

Units:

$$
N \cdot m
$$
