### A car that is stopped starts with an acceleration of 1.5 m/s2. At that same moment it is overtaken by a truck that has a constant speed of 15 m/s. Calculate the meeting position of both vehicles

```python
from Libaldomath import *
from lib_Mechanics import *
```

```python
A=phybody(x1=0,ax=1.5,ang=0,v1=0)
```

```python
B=phybody(x1=0,ax=0,ang=0,v1=15,v2=15)
```

```python
Q=MQ(A.xpos(),B.xpos())
```

```python
t=Q.solve('t','pos')
```

## A particle moves along the X axis with an acceleration a=2cos(t/2) m/s2. At instant t=0, the mobile was at x=-8/2 m, and had speed v=0 m/s.

```python
from Libaldomath import *
from lib_Mechanics import *
```

```python
dx=MyIntg(2*cos(pi*t/2),'dx',var=t)
```

```python
dx.doit('C')
```

```python
C1=dx.solve('C1',t=0,dx=0)
```

```python
dx.set(C1)
```

```python
x=dx.integral('x')
```

```python
x.doit('C')
```

```python
C1=x.solve('C1','update',x=-8/(pi*pi),t=0)
```
