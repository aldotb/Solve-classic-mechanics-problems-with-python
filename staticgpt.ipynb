```python
# iniciando modulo
P=mparticle()
```

```python
# agregando fuerzas que actuan y la direccion
P.add_forza(m*g,-pi/2) # peso
P.add_forza(F,alpha)
P.add_forza(fr,pi)# rozamiento en direccion contraria
P.add_forza(N1,pi/2) # la normal
```

```python
# resultante en x 
Fx=P.Fx('Fx') # retorna un objeto tipo MyEQ
# >> Fx = Fcos(alpha)-fr
```

```python
# resultante en y 
Fy=P.Fy('Fy') # retorna un objeto tipo MyEQ
# >> Fy = Fsin(alpha)-N1 - gm
```

```python
# sabemos que Fy=0 entonces:
N1= Fy.solve(N1)
# c N1=-Fsin(alpha)+gm
```

```python
# en Fx hay movimiento luego 
Q=MQ(a,Fx()*m) # MQ(A,B) or MyEqEq(A,B) : return equation obj  as A = B
# >> a = m*(Fcos(alpha)-fr)
```
