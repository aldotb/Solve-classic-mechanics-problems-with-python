 
# MyVector — 3D Vector Class

A comprehensive **3D vector class** designed for physics, engineering, and advanced mathematics.

It provides a unified framework for vector algebra, geometry, and physics applications, with full symbolic computation support.

## ✨ Features

• Multiple construction methods (components, points, forces, symbolic objects)  
• Complete vector algebra (addition, subtraction, scaling, dot and cross products)  
• Geometric operations (projections, angles, rotations, direction cosines)  
• Physics-oriented methods (torque, moments about axes, force modeling)  
• Full symbolic computation support via SymPy  
• Integrated unit vector handling and coordinate system management  

---

## 🏗 Construction Options

**1. From Components**

`````python
MyVector(1, 2, 3)
MyVector([1, 2, 3])
`````

**2. From Two Points**

`````python
MyVector(Point3D(0,0,0), Point3D(1,2,3))
`````

**3. As a Force Vector**

`````python
MyVector([1,1,0], F=10, origin=(1,2,3))
`````

**4. With Direction and Magnitude**

`````python
MyVector(dire=(1,0,0), F=5)
`````

**5. From SymPy Objects**
 - MyVector(sympy_vector)
 - MyVector(sympy_matrix)

## ➕ Algebraic Operators

'v + w → Vector addition

`v - w` → Vector subtraction

`3 * v` → Scalar multiplication

`v * 3` → Scalar multiplication

`v.dot(w)` → Dot product (returns scalar)

`v @ w` → Dot product using `@` operator

`v ** w` → Cross product

`v | w` → Angle between vectors

---

## 📐 Properties

`v.components` → $(x, y, z)$

`v.magnitude` → $|\vec{v}|$

`v.unit_vector` → $\dfrac{\vec{v}}{|\vec{v}|}$

`v.origin` → Application point

`v.endpoint` → Terminal point

`v.direction_cosines` → $(\cos\alpha, \cos\beta, \cos\gamma)$

---

## ⚙ Physics Methods

`v.torque(reference_point)`

Returns the torque vector:

$$
\vec{\tau} = \vec{r} \times \vec{F}
$$

`v.moment_about_axis(point, axis)`

Returns the scalar moment about a specified axis.

`v.projection(other)`

Returns the vector projection onto another vector:

$$
\text{proj}_{\vec{w}} \vec{v} =
\frac{\vec{v}\cdot\vec{w}}{|\vec{w}|^2} \vec{w}
$$
