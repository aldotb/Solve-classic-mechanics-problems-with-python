Particle Mechanics with lib_Physics and mparticle python class

This notebook introduces the particle mechanics utilities provided by
the `lib_Physics` module.

The goal of the library is to simplify the formulation of Newton's laws
using symbolic computation with SymPy.

````python
from lib_Physics import * 
````

Particle Model

A particle is created using the `mparticle()` class.

The particle stores:

- mass
- forces acting on the particle
- symbolic equations of motion

````python
P = mparticle()
````

````python
P.m
````

Adding Forces

Forces are added using the function:

```
add_forza(value, angle)
```

where:

`value` = magnitude of the force  
`angle` = direction measured from the positive x-axis

````python
P.add_forza(m*g,-pi/2) # weight
P.add_forza(F1,0)
P.add_forza(fr,pi)# rozamiento en direccion contraria
P.add_forza(N1,pi/2) # la normal
````

Force Components

If a force has magnitude $F$ and direction $\theta$:

$$
F_x = F\cos(\theta)
$$

$$
F_y = F\sin(\theta)
$$

The library automatically computes these components.

Forces Applied at a Specific Point

A force can also be applied at a specific point:

```
add_forza(value,angle,x,y)
```

This is useful when computing moments or torques.

````python
P.add_forza(F,theta,x,y)
````

Moment Generated by a Force

If the force is applied at point $(x,y)$, the moment about the origin is:

$$
M = xF_y - yF_x
$$

Resultant Forces

The total forces acting on the particle can be obtained with:

````python
P.Fx()
````

````python
P.Fy()
````

Physics Interpretation

These correspond to the sums:

$$
\sum F_x
$$

$$
\sum F_y
$$

Equations of Motion

The equations of motion are obtained using:

````python
P.xres()
````

````python
P.yres()
````

Newton's Second Law

The equations correspond to:

$$
\sum F_x = m a_x
$$

$$
\sum F_y = m a_y
$$

Summary

The workflow implemented in `lib_Physics` is:

```
mparticle()
      ↓
add_forza()
      ↓
Fx(), Fy()
      ↓
xres(), yres()
      ↓
solve with SymPy
```

This allows a free-body diagram to be translated directly into symbolic
equations of motion.
