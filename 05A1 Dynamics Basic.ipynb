 

# Dynamics — Basic Theory

## 1. Introduction

Dynamics studies the **causes of motion**.
While kinematics describes motion without considering forces, dynamics relates motion to the **forces acting on bodies**.

The foundation of classical dynamics is given by **Newton’s laws of motion**, which establish the relationship between force, mass, and acceleration. ([Encyclopedia Britannica][1])

---

# 2. Newton's Laws of Motion

## First Law — Law of Inertia

A body remains at rest or moves with constant velocity unless acted upon by a net external force.

Mathematically:

$$
\sum \vec{F} = 0 \quad \Rightarrow \quad \vec{v} = \text{constant}
$$

This law defines **inertial reference frames**, where Newtonian mechanics is valid.
