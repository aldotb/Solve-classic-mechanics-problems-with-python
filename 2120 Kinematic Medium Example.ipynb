```python
# y(2*L) = 3*L....then create a MyEq equation to manipulate
Q1=MyEq(3*L-y(2*L),'Q1')
```

```python
Q1.reducecero() # if Q = k*(x+1) and Q=0..then Q=(x+1) i hope tyou know algebra
```

```python
Q1.simplify()
```

```python
V=Q1.solve('V','positive')
```

```python
Q2.reducecero()
```

```python
# in this step we change V for the value finding when call solve('V')...
Q2.set(V)
```
