```python
# from upper answer when center is x1,y1 and R=R then.... CG = x1, y1+2*R/pi...
```

```python
# from upper answer when center is x1,y1 and R=R then.... CG = x1, y1+2*R/pi...
def cgc(x1,y2,R,op='up'):
    mm=pi*R
    cx=x1
    if op=='down': # if refer to down semicircle  
        cy=y2-2*R/pi  
    else:
        cy=y2+2*R/pi
    return (mm,cx,cy)
```

```python
P=phybody()
```

```python
cgc(R/2,0,R/2)
```

```python
P.add_cg(*cgc(R/2,0,R/2))
```

```python
cgc(3*R/2,0,R/2,'down')
```

```python
P.add_cg(*cgc(3*R/2,0,R/2,'down'))
```

```python
cgc(R,0,R,'down')
```

```python
P.add_cg(*cgc(R,0,R,'down'))
```

```python
# join together
P.get_cg()
```
